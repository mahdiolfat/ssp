```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

```python
def lapace():
	s = np.linspace(-2 * np.pi, 0, 100, endpoint=False)
	pdf = 0.5 * np.exp(-np.abs(s))
	pdf = np.concatenate((pdf, [0.5], np.flip(pdf)))
	s = np.concatenate((s, [0], -np.flip(s)))
	return pdf, s
```

```python
pdf, s = lapace()
plt.plot(s, pdf)
```

```python
def exponential(M=11, decay=2):
	s = np.linspace(0, 2 * np.pi, M)
	pdf = np.exp(-decay * s)
	return pdf, s

pdf, s = exponential()
plt.plot(s, pdf)
```

```python
def rayleigh(M=111, alpha=0.8):
	s = np.linspace(0, 2 * np.pi, M)
	pdf = np.exp(-s**2 / 2 / alpha**2) * s / alpha**2
	return pdf, s

pdf, s = rayleigh()
plt.plot(s, pdf)
```

```python
def kdistribution(M=111, b=1, v=0.9):
	# b is the scale parameter
	s = np.linspace(0, 2 * np.pi, M) + 0.00001
	c = np.sqrt(np.pi / 4 * b)
	gammy = sp.special.gamma(s)
	bessy = sp.special.kv(v - 1, 2 * c * s)
	pdf = 4 * c / gammy * np.power(c * s, v) * bessy
	return pdf, s

pdf, s = kdistribution()
plt.plot(s, pdf)
```
