```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

```python
def parzen(M=11):
	M2 = (M - 1) // 2
	s1 = np.linspace(-M2, -M2 // 2 + 1, M2 - 1) / M2
	s2 = np.linspace(-M2 // 2 + 1, M2 // 2, M2 - 1) / M2
	s3 = np.linspace(M2 // 2, M2, M2 - 1) / M2
	p1 = 2 * np.power((1 - np.abs(s1)), 3)
	p2 = 1 - 6 * np.power(s2, 2) + 6 * np.power(np.abs(s2), 3)
	p3 = 2 * np.power((1 - np.abs(s3)), 3)
	print(s1, s2, s3)
	win = np.concatenate((p1, p2, p3))
	support = np.concatenate((s1, s2, s3))
	return win, support
```

```python
def priestley(M=13):
	M2 = (M - 1) // 2
	s = np.linspace(-M2, 0, M2 // 2, endpoint=False) / M2
	win = 3 / (np.pi * s) ** 2 * (np.sinc(s) - np.cos(np.pi * s))
	win = np.concatenate((win, [1], np.flip(win)))
	return win, s
```

```python
win, _ = priestley(111)
plt.plot(win)
```

```python
def sasaki(M=11):
	M2 = (M - 1) // 2
	swin = np.linspace(-M2, M2, M) / M2
	win = 1 / np.pi * np.abs(np.sin(np.pi * swin)) + (1 - np.abs(swin)) * np.cos(np.pi * swin)
	return win, swin
```

```python
win, _ = sasaki(111)
plt.plot(win)
```
