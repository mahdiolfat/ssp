# Autoregressive Moving Average (ARMA) Process

$$

H(z) = \frac{B_q(z)}{A_p(z)} =
\frac
{
	\displaystyle\sum_{k = 0}^{q}b_q(k)z^{-k}
}
{
	1 + \displaystyle\sum_{k = 1}^{p}a_p(k)z^{-k}
}

$$

## AR Process

$$

H(z) = 
	\frac
	{
		b(0)
	}
	{
		1 + \displaystyle \sum_{k = 1}^{p}a_p(k)z^{-k}
	}

$$

# MA Process

$$

H(z) = {
	\displaystyle\sum_{k = 0}^{q} {
		b_q(k)z^{-k}
	}
}

$$

## Example 3.6.1 An ARMA(2, 2) Random Process

```python
zeros = [0.95 * np.exp(+1j * np.pi / 2),
         0.95 * np.exp(-1j * np.pi / 2)]

poles = [0.9 * np.exp(+1j * 2 * np.pi / 5),
         0.9 * np.exp(-1j * 2 * np.pi / 5)]

rproc = ARMA(poles, zeros, 256)
plt.plot(rproc)
```

$$

H(z) = \frac
{1 + 0.9025z^{-2}}
{1 - 0.5562z^{-1} + 0.81z^{-2}}

$$

```python
fspec = np.fft.rfft(rproc, 1024)
plt.plot(10 * np.log10(np.abs(fspec)))
```
