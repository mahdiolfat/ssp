```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Nwin = 11
Nfft = 64
Nzp = Nfft - Nwin
rect = np.ones(Nwin) / np.sqrt(Nwin)
zph_win = np.pad(rect, (Nzp//2, Nzp//2+1), constant_values=0)
plt.plot(zph_win)
plt.figure()


fft = np.fft.fft(zph_win, norm='ortho')
#plt.plot(np.fft.fftshift(20*np.log10(np.abs(fft))))
l2p = np.fft.fftshift(np.abs(fft)**2)
plt.plot(l2p)
lp2mag = np.sum(l2p)
print(lp2mag)
plt.show()
```
